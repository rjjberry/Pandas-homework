```python
# Overall passing rate
student_overall_passing_rate = (student_data_df['math_score'].mean() + student_data_df['reading_score'].mean()) / 2
```

```python
# Percentage with passing Math Score(>=70)

#student_math_pass_count = student_data_df['math_score'] >= 70
#student_math_pass_count = (student_data_df['math_score'] >= 70).value_counts
#student_math_pass_count = pd.value_counts(student_data_df['math_score'] >= 70)
# student_math_pass_count2 = student_math_pass_count == True
#student_math_pass_count.unique()

#student_math_pass_count = student_data_df.math_score >= 70
#student_math_pass_count

student_total_math_score_count = student_data_df['math_score'].count()

student_math_pass_count_prep = student_data_df[student_data_df.math_score >= 70]
student_math_pass_count = student_math_pass_count_prep.math_score.count()
student_math_pass_count

student_math_pass_pct = (student_math_pass_count / student_total_math_score_count) * 100
```

```python
# Percentage with passing Reading Score(>=70)

student_total_reading_score_count = student_data_df['reading_score'].count()

student_reading_pass_count_prep = student_data_df[student_data_df.reading_score >= 70]
student_reading_pass_count = student_reading_pass_count_prep.reading_score.count()
student_reading_pass_count

student_reading_pass_pct = (student_reading_pass_count / student_total_reading_score_count) * 100
```
